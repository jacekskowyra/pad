```python

# use pandas to extract rainfall inches as a NumPy array
rainfall = pd.read_csv('Seattle2014.csv')['PRCP'].values
inches = rainfall / 254.0  # 1/10mm -> inches
inches.shape
print(inches)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # set plot styles
plt.hist(inches, 40);
```

```python
zeros = np.count_nonzero(inches==0)
print("Number days without rain:    ",zeros)
```

```python
non_zeros = np.count_nonzero(inches!=0)
print("Number days with rain:       ",non_zeros)
```

```python
more_than_05 = np.count_nonzero(inches>0.5)
print("Days with more than 0.5 inches. :",more_than_05)
```

```python
less_than_02 = np.count_nonzero((0 < inches) & (inches < 0.2))


print("Rainy days with < 0.2 inches.  :",less_than_02)
```
