```python
import pandas as pd
```

```python
orders=pd.read_csv("orders.csv")

orders.describe()
```

```python
orders.head()
```

```python
orders.info()
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'], format='%Y/%m/%d')

orders.head()
```

```python
print(orders['tshirt_category'].unique())
```

```python
orders['tshirt_category'].nunique()
```

```python
#orders['tshirt_category'] = orders['tshirt_category'].replace("Wh ","White ",regex=True)
#orders['tshirt_category'] = orders['tshirt_category'].replace("Tshirt","T-Shirt",regex=True)

orders["tshirt_category"] = orders["tshirt_category"].apply(lambda x: x.replace("Wh ", "White "))
orders["tshirt_category"] = orders["tshirt_category"].apply(lambda x: x.replace("Tshirt", "T-Shirt"))
orders["tshirt_category"] = orders["tshirt_category"].apply(lambda x: x.replace("Bl ", "Black "))




```

```python
print(orders['tshirt_category'].unique())
```

```python
orders[['tshirt_color', 'tshirt_type', 'tshirt_size']] = orders['tshirt_category'].apply(lambda x: pd.Series(str(x).split(" ")))
```

```python
orders.head()
```

```python
filtered_dates = orders.loc[(orders['order_date'] >= '2014-12-31') & (orders['order_date'] < '2016-08-02')]
filtered_dates
```
